# Molar Volume



While none of the values found in molar volumes are non unphysical, some are very high, so we should look at the high values and verify them

```python
plot('Molar_volume')
```

```python
odd_molar_vol = df[df['Molar_volume']>1000]

odd_molar_vol[['SMILES_Monomer','Molar_volume','Common_names']]
```

It looks liek the absurdly high values come from a citation on the web scraping. To correct this, we can just drop the 10's value 

```python
corrected = (odd_molar_vol['Molar_volume'] *0.1)-0.1
corrected
```

```python
odd_molar_vol['Molar_volume'] = corrected
odd_molar_vol['Molar_volume']
```

```python
df.loc[odd_molar_vol.index]= odd_molar_vol
```

```python
# test
plot('Molar_volume')
```
