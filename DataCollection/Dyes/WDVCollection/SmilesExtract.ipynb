# Extract the CAS numbers from structured text

```python
# make a small function to extract CAS number from the text
def getCAS(txt):
    try: 
        casnum = re.search('CAS Registry Number:(.*)',txt)
        cas = casnum.groups()[0]
    except:
        cas = None
    return cas
```

```python
DyeData['CASNumber'] = DyeData['CAS'].apply(getCAS)
```

```python
DyeData['CAS'] = DyeData['CASNumber']
DyeData=DyeData.drop(['CASNumber'],axis=1)
```

```python
# make a quick function to get the SMILES of each molecule
# for CAS ref
def resolve(x):
    try:
        z = cirpy.resolve(x, 'smiles')
    except:
        z = None
    return z 
```

```python
DyeData['SMILES'] = DyeData['CAS'].progress_apply(resolve)
```
