# make function for doin this to all data points

```python
Cols = ['name','CAS','LF','Standard','Dyetype','class','properties']

def structure(file, Dyetype):
    f =open(file)
    DyeData = json.load(f)
    Data = []
    for i in range(0,len(DyeData)):
        try:
            data = DyeData[i]
            v = data['verbal']
            n = data['numerical']
            CAS = v['CAS']
            LF = n[2][2]
            Standard = n[2][0]
            Class = v['Mol']
            properties = v['properties']
            name = v['name']

            Data.append([name,CAS,LF,Standard, Dyetype, Class, properties])
        except:
            next
    df = pd.DataFrame(Data, columns=Cols)
    return df
```

```python
Aciddf = structure('/home/matt/Proj/Dyes/Dyebrary/DyeScraping/acidDyes.json','AcidDye')
Azoicdf = structure('/home/matt/Proj/Dyes/Dyebrary/DyeScraping/azoicDyes.json','AzoicDye')
Basicdf = structure('/home/matt/Proj/Dyes/Dyebrary/DyeScraping/basicDyes.json','BasicDye')
Directdf = structure('/home/matt/Proj/Dyes/Dyebrary/DyeScraping/directDyes.json','DirectDye')
Dispersedf = structure('/home/matt/Proj/Dyes/Dyebrary/DyeScraping/disperseDyes.json','DisperseDye')
foodDF = structure('/home/matt/Proj/Dyes/Dyebrary/DyeScraping/foodDyes.json','FoodDye')
Mordantdf = structure('/home/matt/Proj/Dyes/Dyebrary/DyeScraping/mordantDyes.json','MordantDye')
Reactivedf = structure('/home/matt/Proj/Dyes/Dyebrary/DyeScraping/reactiveDyes.json','ReactiveDye')
Solventdf = structure('/home/matt/Proj/Dyes/Dyebrary/DyeScraping/solventDyes.json','SolventDye')
Sulfurdf = structure('/home/matt/Proj/Dyes/Dyebrary/DyeScraping/sulfurDyes.json','SulfurDye')
Vatdf = structure('/home/matt/Proj/Dyes/Dyebrary/DyeScraping/vatDyes.json','VatDye')
```
